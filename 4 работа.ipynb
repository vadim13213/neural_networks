```python
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

""" 1.    """
def create_gantt_chart():
    #   
    tasks = [
        {"Task": "  ", "Start": "2025-03-01", "End": "2025-03-14", "Group": ""},
        {"Task": " ", "Start": "2025-03-01", "End": "2025-03-05", "Group": ""},
        {"Task": " ", "Start": "2025-03-06", "End": "2025-03-09", "Group": ""},
        {"Task": " ", "Start": "2025-03-10", "End": "2025-03-14", "Group": ""},
        {"Task": " ", "Start": "2025-03-15", "End": "2025-04-04", "Group": ""},
        {"Task": "  ", "Start": "2025-03-15", "End": "2025-03-21", "Group": ""},
        {"Task": " ", "Start": "2025-03-15", "End": "2025-03-21", "Group": ""},
        {"Task": " API", "Start": "2025-03-22", "End": "2025-03-28", "Group": ""},
        {"Task": " ", "Start": "2025-03-22", "End": "2025-03-28", "Group": ""},
        {"Task": "", "Start": "2025-04-05", "End": "2025-05-09", "Group": ""},
        {"Task": "  Arduino", "Start": "2025-04-05", "End": "2025-04-18", "Group": ""},
        {"Task": " ", "Start": "2025-04-05", "End": "2025-04-18", "Group": ""},
        {"Task": " ", "Start": "2025-04-05", "End": "2025-04-25", "Group": ""},
        {"Task": " ", "Start": "2025-04-26", "End": "2025-05-02", "Group": ""},
        {"Task": " ", "Start": "2025-05-10", "End": "2025-05-24", "Group": ""},
        {"Task": " ", "Start": "2025-05-10", "End": "2025-05-14", "Group": ""},
        {"Task": " ", "Start": "2025-05-15", "End": "2025-05-19", "Group": ""},
        {"Task": " ", "Start": "2025-05-20", "End": "2025-05-24", "Group": ""},
        {"Task": "", "Start": "2025-05-25", "End": "2025-05-31", "Group": ""},
        {"Task": " ", "Start": "2025-05-25", "End": "2025-05-28", "Group": ""},
        {"Task": " ", "Start": "2025-05-29", "End": "2025-05-31", "Group": ""}
    ]

    #  DataFrame
    df = pd.DataFrame(tasks)
    df['Start'] = pd.to_datetime(df['Start'])
    df['End'] = pd.to_datetime(df['End'])
    df['Duration'] = (df['End'] - df['Start']).dt.days + 1

    #    
    colors = {
        '': '#FFD700',
        '': '#87CEEB',
        '': '#98FB98',
        '': '#FFA07A',
        '': '#DDA0DD'
    }

    #  
    fig, ax = plt.subplots(figsize=(18, 10))

    #  
    for i, task in enumerate(df.itertuples()):
        ax.barh(y=task.Task,
                width=task.Duration,
                left=task.Start,
                color=colors[task.Group],
                edgecolor='black',
                height=0.6,
                label=task.Group)

        #  
        if task.Duration > 3:
            ax.text(x=task.Start + timedelta(days=task.Duration/2),
                    y=i,
                    s=f"{task.Duration}",
                    va='center',
                    ha='center',
                    color='black',
                    fontsize=9)

    #  
    ax.set_title(' :    Arduino\n', fontsize=16, pad=20)
    ax.set_xlabel('', fontsize=12, labelpad=15)
    ax.set_ylabel(' ', fontsize=12, labelpad=15)

    #    
    ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MO))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
    ax.xaxis.set_minor_locator(mdates.DayLocator())

    plt.xticks(rotation=45, ha='right')
    plt.grid(which='both', axis='x', linestyle=':', color='gray', alpha=0.7)
    plt.grid(which='major', axis='x', linestyle='-', color='black', alpha=0.5)

    # 
    handles = [plt.Rectangle((0,0),1,1, color=colors[g]) for g in colors]
    ax.legend(handles, colors.keys(), title=' ', loc='upper right')

    plt.tight_layout()
    plt.savefig('microfarm_gantt.png', dpi=300, bbox_inches='tight')
    plt.show()

""" 2.    """
def generate_risks_table():
    risks = [
        {"ID": "R1", "": " ", "": "", "": "", "": "", "": ",  "},
        {"ID": "R2", "": " ", "": "", "": "", "": "", "": "Wi-Fi + Bluetooth, "},
        {"ID": "R3", "": "  ", "": "", "": "", "": "", "": "  "},
        {"ID": "R4", "": " ", "": "", "": "", "": "", "": " "},
        {"ID": "R5", "": " ", "": "", "": "", "": "", "": " "}
    ]

    df_risks = pd.DataFrame(risks)
    df_risks.to_excel('project_risks.xlsx', index=False)

""" 3.    """
def generate_milestones():
    milestones = [
        {"": 1, "": " ", "": "14.03.2025", "": "  "},
        {"": 2, "": " ", "": "04.04.2025", "": "  "},
        {"": 3, "": " ", "": "09.05.2025", "": " "},
        {"": 4, "": " ", "": "24.05.2025", "": "95%  "},
        {"": 5, "": " ", "": "31.05.2025", "": "  "}
    ]

    df_milestones = pd.DataFrame(milestones)
    df_milestones.to_excel('project_milestones.xlsx', index=False)

""" 4.    """
def generate_report():
    report = """#  

## 1.  
  5 :
1.  (14 ) -  
2.  (21 ) -  
3.  (35 ) -  
4.  (14 ) -  
5.  (7 ) - 

## 2.  
 :
-   (: )
-   (: )
 :
-  
-   

## 3.  
 :
1. 14.03.2025 -  
2. 04.04.2025 -  
3. 09.05.2025 -  

          .
"""
    with open('project_report.txt', 'w', encoding='utf-8') as f:
        f.write(report)

"""    """
if __name__ == "__main__":
    create_gantt_chart()
    generate_risks_table()
    generate_milestones()
    generate_report()
    print("   :")
    print("- microfarm_gantt.png")
    print("- project_risks.xlsx")
    print("- project_milestones.xlsx")
    print("- project_report.txt")
```
